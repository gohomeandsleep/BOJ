```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import matplotlib.font_manager as fm
fontpath = "/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf"
font = fm.FontProperties(fname = fontpath, size = 9)
plt.rc('font', family = "NanumBarunGothic")
mpl.font_manager._rebuild()
df=pd.read_csv('/content/sample_data/Score.csv')
df.plot()
plt.title('중간고사 개인별 성적 현황 분석')
plt.xlabel('학생이름')
plt.ylabel('합계점수')
plt.xticks(range(10), labels=df['이름'].head(10), rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (8, 5))
sns.barplot(x='이름', y='국어', data=df.head(10))
plt.title('중간고사 개인별 국어점수 현황')
plt.xlabel('이름')
plt.ylabel('국어')
plt.show()
```

```python
grade = sorted(df['분반'].unique())
plt.figure(figsize=(8,4))
sns.countplot(x='분반', hue='성별',data=df, order = grade)
plt.title('중간고사 반별 응시인원수 현황')
plt.show()
```
